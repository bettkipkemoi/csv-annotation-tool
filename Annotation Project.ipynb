```python
df = pd.read_excel('rows_to_label.xlsx')
```

```python
df.head(10)
```

```python
def annotate_helplessness(text):
    count = 0
    # define patterns for learned helplessness
    patterns = ['feel helpless', 'no passion', 'no motivation', 'overwhelmed', 'i am a failure', 'won\'t', 'feel powerless', 'wish', 'no control', 'always fail', 'dissapointed', 'poorly', 'no way out', 'stuck', 'hopeless', 'no hope', 'lonely', 'pity']
    for pattern in patterns:
        if pattern in text:
            count += 1
    if count == 0:
        return 1 # no presence
    elif count == 1:
        return 2 # low presence
    elif count == 2:
        return 3 # moderate presence
    else:
        return 4 # high presence
```

```python
df['sage_label'] = df['Body'].apply(annotate_helplessness)
```
